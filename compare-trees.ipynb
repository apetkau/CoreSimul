```r
library(ape)
library(phytools)
library(phangorn)
library(repr)

t_orig <- read.tree(file="example.tree")
t_new <- read.tree(file="test/sequences/core.aln.treefile")
t_renamed <- read.tree(file="test/renamed.tree")

t_orig <- root(t_orig, "B16", resolve.root=TRUE)
t_new <- root(t_new, "B16", resolve.root=TRUE)
t_renamed <- root(t_renamed, "BOZO0034", resolve.root=TRUE)

comparePhylo(t_orig, t_new, plot=TRUE)
```

```r
treedist(t_orig, t_new, check.labels = TRUE)
```

```r
options(repr.plot.width=10, repr.plot.height=8)

plot(cophylo(t_orig, t_new, rotate=TRUE))
```

```r
tip_association <- read.table("test/names.txt")
tip_association["tmp"] <- tip_association[,1]
tip_association["V1"] <- tip_association["V2"]
tip_association["V2"] <- tip_association["tmp"]
tip_association["tmp"] <- NULL
tip_association
```

```r
plot(cophylo(t_orig, t_renamed, rotate=TRUE, assoc=tip_association))
```
